```python
token_list = []
def tokenization(law):
    tokens = word_tokenize(law)
    token_list.append(tokens)
    return token_list
```

```python
def remove_punctuations(word_list):
    withoutPunctuationLaws = []
    
    for words in word_list:
        
        withoutPunctuation = []
        
        for token in words:    
            if token.isalpha():
                if len(token)>0:
                    withoutPunctuation.append(token)
                
        withoutPunctuationLaws.append(withoutPunctuation)
    
    return withoutPunctuationLaws
```

```python
def remove_stopwords(word_list):
    stop_words = set(stopwords.words("english"))
    swRemovedList = []
    
    for words in word_list:
        withoutStopWords = []
        
        for token in words:
            if (isinstance(token, str)):
                if token.lower() not in stop_words:
                    withoutStopWords.append(token)
        swRemovedList.append(withoutStopWords)
    return swRemovedList
```

```python
def remove_duplicate_words(word_list):
    
    unique_words = []

    for words in word_list:
        
        unique_words_in_law = []
        seen_words = set()

        for token in words:
            # Convert the word to lowercase for case-insensitive comparison
            normalized_word = token.lower()
        
            if normalized_word not in seen_words:
                unique_words_in_law.append(normalized_word)
                seen_words.add(normalized_word)
                
        unique_words.append(unique_words_in_law)
        
    return unique_words
```

```python
pstemmer = PorterStemmer()

def word_stemming(word_list):
    stemmed_list = []
    for words in word_list:
        stemed_word_in_law = []
        for tokens in words:
            stem = pstemmer.stem(tokens)
            stemed_word_in_law.append(stem)
        stemmed_list.append(stemed_word_in_law)
    return stemmed_list
```

```python
input_text = """The law allows for the application of special measures to protect vulnerable victims of crime or witnesses during legal proceedings. These measures are intended to safeguard their well-being and ensure a fair trial."""
input_list = []
input_tokens = word_tokenize(input_text)
input_list.append(input_tokens)

removed_pun_list = remove_punctuations(input_list)
filtered_tokens = remove_stopwords(removed_pun_list)
unique_filtered_tokens_input = remove_duplicate_words(filtered_tokens)
#input_features = word_stemming(unique_filtered_tokens)

unique_filtered_tokens_input
```

```python
import numpy as np

cosine_similarity_list = []

# Assuming A and B are your 300-dimensional vectors
A = input_vector

for vectors in all_laws_vectors:
    B = vectors

    # Calculate dot product
    dot_product = np.dot(A, B)

    # Calculate Euclidean norms
    norm_A = np.linalg.norm(A)
    norm_B = np.linalg.norm(B)

    # Calculate cosine similarity
    cosine_similarity = dot_product / (norm_A * norm_B)
    cosine_similarity_list.append(cosine_similarity)


print("Cosine Similarity list: ", cosine_similarity_list)
```
