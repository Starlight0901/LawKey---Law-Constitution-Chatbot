```python
def law_cleaning(law):
    # Remove punctuations
    law = law.translate(str.maketrans('', '', string.punctuation + '\r\n\t'))
    
    # Remove special characters
    law = law.replace('ã', '')
    law = law.replace('Ã', '')
    
    # remain only the alabetic, numeric characters and whitespaces
    law = ''.join([i for i in law if i.isalnum() or i.isspace()])
    return law
```

```python
def tokenize(text):
    tokens = nltk.word_tokenize(text.lower())
    return tokens
```

```python
def remove_stopwords(tokens):
    stop_words = set(stopwords.words("english"))
    
    filtered = []
    
    for token in tokens:
        if token.lower() not in stop_words:
            filtered.append(token)
    
    return filtered
```

```python
def stemmer(tokens):
    # creaate the porter stemmer object
    pstemmer = nltk.PorterStemmer()
    
    stemmed_words = []
    for token in tokens:
        stemmed_word = pstemmer.stem(token)
        stemmed_words.append(stemmed_word)
    
    # Join the list of stemmed words into a single string separated by spaces
    stemmed_text = ' '.join(stemmed_words)

    return stemmed_text
```

```python
df["law_clean"] = df["Law"].apply(law_cleaning)
df["tokens"] = df["law_clean"].apply(tokenize)
df["tokens"] = df["tokens"].apply(remove_stopwords)
df["tokens"] = df["tokens"].apply(stemmer)
df["tokens_w"] = df["tokens"].apply(tokenize)
```

```python
def get_mean_vectors(features):
    tokens = []
    for token in features:
        if token in model:
            tokens.append(token)
    if len(tokens) >= 1:
        return model.get_mean_vector(tokens)
    else:
        return []
```

```python
def retrieve_laws(input_text):
    tokens = nltk.word_tokenize(input_text)
    tokens = [token for token in tokens if token.lower() not in stopwords.words('english')]
    tokens = [nltk.PorterStemmer().stem(word) for word in tokens]
    
    # convert into vectors
    input_vectors = get_mean_vectors(tokens)
    
    similarity_level = cosine_similarity([input_vectors], df['Vectors'].tolist())[0]
    df['similarity_to_input'] = similarity_level
    
    # Get the top 1 based on cosine similarity
    related_laws = df.sort_values(by='similarity_to_input', ascending=False).head(1)['Law'].tolist()
    key = df.sort_values(by='similarity_to_input', ascending=False).head(1)['Key'].tolist()
    print(key)
    
    return related_laws
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix


# Function to retrieve the predicted law for a given input
def get_predicted_law(input_text):
    related_laws = retrieve_laws(input_text)
    if related_laws:
        return related_laws[0]
    else:
        return None  # Handle cases where no similar law is found

# Initialize variables for evaluation metrics
correct_predictions = 0
total_predictions = 0
true_positives = 0
false_positives = 0
false_negatives = 0

# Iterate through each sample input
predicted_laws_list = []
for index, row in testing_df.iterrows():
    input_text = row["label questions"]
    actual_law = row["Law"]

    predicted_law = get_predicted_law(input_text)
    predicted_laws_list.append(predicted_law)

    # Check if the prediction is correct
    if predicted_law == actual_law:
        correct_predictions += 1
        true_positives += 1
    else:
        false_positives += 1  # Incorrect prediction
        false_negatives += 1  # Missed the correct law

    total_predictions += 1

# add the predicted laws to the dataframe
testing_df["predicted_laws"] = predicted_laws_list

    
# Calculate accuracy, precision, recall, and F1-score
accuracy = accuracy_score(testing_df["Law"], testing_df["predicted_laws"])
precision = precision_score(testing_df["Law"], testing_df["predicted_laws"], average='macro')
recall = recall_score(testing_df["Law"], testing_df["predicted_laws"], average='macro')
f1_score = f1_score(testing_df["Law"], testing_df["predicted_laws"], average='macro')

# Print evaluation metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1_score)

```
